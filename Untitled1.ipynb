```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from sklearn.linear_model import LinearRegression
import io

# Function to upload Excel sheet
def upload_file():
    uploaded = files.upload()
    for fn in uploaded.keys():
        data = pd.read_excel(io.BytesIO(uploaded[fn]))
    return data

# Upload the dataset
print("DS.xlsx(application/vnd.openxmlformats-officedocument.spreadsheetml.sheet) - 19907 bytes, last modified: 12/15/2024 - 100% done")
print("Saving DS.xlsx to DS (5).xlsx")
data = upload_file()

# Check the first few rows of the dataset
print(data.head())

# Extract necessary columns for analysis
promotion_data = data[['Date', 'Sales', 'Monthly Promotion']].copy()

# Convert Date to datetime format
promotion_data['Date'] = pd.to_datetime(promotion_data['Date'])

# Group by month to aggregate sales and promotions
promotion_data['Month'] = promotion_data['Date'].dt.to_period('M')
monthly_data = promotion_data.groupby('Month').agg({'Sales': 'sum', 'Monthly Promotion': 'sum'}).reset_index()

# Separate sales with and without promotions
monthly_data['Sales with Promotion'] = promotion_data[promotion_data['Monthly Promotion'] == 0].groupby('Month')['Sales'].sum().values
monthly_data['Sales without Promotion'] = promotion_data[promotion_data['Monthly Promotion'] == 1].groupby('Month')['Sales'].sum().values

# Fill any missing values with 0
monthly_data = monthly_data.fillna(0)

# Plot the impact of monthly promotions on sales
plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Month'].astype(str), monthly_data['Sales with Promotion'], label='Sales with Promotion')
plt.plot(monthly_data['Month'].astype(str), monthly_data['Sales without Promotion'], label='Sales without Promotion')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Impact of Monthly Promotions on Sales')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Predict future sales using linear regression for both with and without promotion
X = monthly_data[['Monthly Promotion']]
y_with_promotion = monthly_data['Sales with Promotion']
y_without_promotion = monthly_data['Sales without Promotion']

model_with_promotion = LinearRegression()
model_without_promotion = LinearRegression()

model_with_promotion.fit(X, y_with_promotion)
model_without_promotion.fit(X, y_without_promotion)

# Determine the starting month for the forecast
last_date_in_data = promotion_data['Date'].max()
first_forecast_month = (last_date_in_data + pd.DateOffset(months=1)).to_period('M')

# Forecast for next 5 months
future_promotions = pd.DataFrame({'Monthly Promotion': [0, 1, 1, 0, 1]})  # Dummy promotion data
future_sales_with_promotion = model_with_promotion.predict(future_promotions)
future_sales_without_promotion = model_without_promotion.predict(future_promotions)

# Generate future dates for the forecast
future_dates = pd.date_range(start=first_forecast_month.start_time, periods=5, freq='MS')

# Combine past sales for forecast plot
monthly_data['All Sales'] = monthly_data['Sales with Promotion'] + monthly_data['Sales without Promotion']

# Predict future sales for combined data
model_all_sales = LinearRegression()
model_all_sales.fit(X, monthly_data['All Sales'])
future_all_sales = model_all_sales.predict(future_promotions)

# Plot future forecast
plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Month'].astype(str), monthly_data['Sales with Promotion'], label='Past Sales with Promotion')
plt.plot(monthly_data['Month'].astype(str), monthly_data['Sales without Promotion'], label='Past Sales without Promotion')
plt.plot(future_dates.astype(str), future_sales_with_promotion, label='Forecasted Sales with Promotion', linestyle='--')
plt.plot(future_dates.astype(str), future_sales_without_promotion, label='Forecasted Sales without Promotion', linestyle='--')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales Forecast for Next 5 Months')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Plot forecast for all sales
plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Month'].astype(str), monthly_data['All Sales'], label='Past All Sales')
plt.plot(future_dates.astype(str), future_all_sales, label='Forecasted All Sales', linestyle='--')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('All Sales Forecast for Next 5 Months')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from sklearn.linear_model import LinearRegression
import io

# Function to upload Excel sheet
def upload_file():
    uploaded = files.upload()
    for fn in uploaded.keys():
        data = pd.read_excel(io.BytesIO(uploaded[fn]))
    return data

# Upload the dataset
print("DS.xlsx(application/vnd.openxmlformats-officedocument.spreadsheetml.sheet) - 19907 bytes, last modified: 12/15/2024 - 100% done")
print("Saving DS.xlsx to DS (5).xlsx")
data = upload_file()

# Check the first few rows of the dataset
print(data.head())

# Extract necessary columns for analysis
manufacturer_data = data[['Date', 'Item', 'Sales', 'Manufacturer']].copy()

# Convert Date to datetime format
manufacturer_data['Date'] = pd.to_datetime(manufacturer_data['Date'])

# Group by month and manufacturer to aggregate sales
manufacturer_data['Month'] = manufacturer_data['Date'].dt.to_period('M')
monthly_manufacturer_data = manufacturer_data.groupby(['Month', 'Manufacturer']).agg({'Sales': 'sum'}).reset_index()

# Plot the impact of different manufacturers on sales
plt.figure(figsize=(12, 6))
for manufacturer in monthly_manufacturer_data['Manufacturer'].unique():
    manufacturer_sales = monthly_manufacturer_data[monthly_manufacturer_data['Manufacturer'] == manufacturer]
    plt.plot(manufacturer_sales['Month'].astype(str), manufacturer_sales['Sales'], label=manufacturer)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Impact of Different Manufacturers on Sales')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Predict future sales using linear regression for each manufacturer
future_sales = []
future_dates = pd.date_range(start=manufacturer_data['Date'].max(), periods=6, freq='MS')[1:]

for manufacturer in monthly_manufacturer_data['Manufacturer'].unique():
    manufacturer_sales = monthly_manufacturer_data[monthly_manufacturer_data['Manufacturer'] == manufacturer]
    X = np.arange(len(manufacturer_sales)).reshape(-1, 1)
    y = manufacturer_sales['Sales']

    model = LinearRegression()
    model.fit(X, y)

    future_X = np.arange(len(manufacturer_sales), len(manufacturer_sales) + 5).reshape(-1, 1)
    manufacturer_future_sales = model.predict(future_X)

    future_sales.append((manufacturer, manufacturer_future_sales))

# Plot future forecast for each manufacturer
plt.figure(figsize=(12, 6))
for manufacturer, sales in future_sales:
    manufacturer_sales = monthly_manufacturer_data[monthly_manufacturer_data['Manufacturer'] == manufacturer]
    plt.plot(manufacturer_sales['Month'].astype(str), manufacturer_sales['Sales'], label=f'Past Sales - {manufacturer}')
    plt.plot(future_dates.astype(str), sales, label=f'Forecasted Sales - {manufacturer}', linestyle='--')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales Forecast for Next 5 Months by Manufacturer')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Function to generate a random date within the last 8 months
def random_date(start, end):
    return start + timedelta(days=random.randint(0, int((end - start).days)))

# Generating sample data
num_rows = 1000
data = {
    'Branch ID': np.random.randint(1, 10, num_rows),
    'Branch Name': [random.choice(['Branch A', 'Branch B', 'Branch C', 'Branch D', 'Branch E']) for _ in range(num_rows)],
    'Date': [random_date(datetime.now() - timedelta(days=240), datetime.now()) for _ in range(num_rows)],
    'Transaction ID': [f'TX{str(i).zfill(5)}' for i in range(num_rows)],
    'Product ID': [f'P{str(i).zfill(4)}' for i in range(num_rows)],
    'Product Name': [random.choice(['Product X', 'Product Y', 'Product Z', 'Product W']) for _ in range(num_rows)],
    'Product Category': [random.choice(['Groceries', 'Electronics', 'Clothing', 'Stationery']) for _ in range(num_rows)],
    'Quantity Sold': np.random.randint(1, 20, num_rows),
    'Unit Price': np.round(np.random.uniform(10, 500, num_rows), 2),
    'Customer ID': [f'C{str(i).zfill(6)}' for i in range(num_rows)],
    'Payment Method': [random.choice(['Cash', 'Credit Card', 'Debit Card', 'Online']) for _ in range(num_rows)],
    'Discount': np.round(np.random.uniform(0, 50, num_rows), 2),
    'Salesperson ID': [f'S{str(i).zfill(4)}' for i in range(num_rows)],
}

# Converting the data to a DataFrame
df = pd.DataFrame(data)

# Calculating the Total Price
df['Total Price'] = df['Quantity Sold'] * df['Unit Price']

# Adding Week Number, Month, and Year columns
df['Week Number'] = df['Date'].apply(lambda x: x.isocalendar()[1])
df['Month'] = df['Date'].apply(lambda x: x.month)
df['Year'] = df['Date'].apply(lambda x: x.year)

# Saving the DataFrame to an Excel file
df.to_excel('sales_data_sample.xlsx', index=False)

print("Sample Excel file 'sales_data_sample.xlsx' created successfully.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from sklearn.linear_model import LinearRegression
import io

# Function to upload Excel sheet
def upload_file():
    uploaded = files.upload()
    for fn in uploaded.keys():
        data = pd.read_excel(io.BytesIO(uploaded[fn]))
    return data

# Upload the dataset
print("DS.xlsx(application/vnd.openxmlformats-officedocument.spreadsheetml.sheet) - 19907 bytes, last modified: 12/15/2024 - 100% done")
print("Saving DS.xlsx to DS (5).xlsx")
data = upload_file()

# Check the first few rows of the dataset
print(data.head())

# Extract necessary columns for analysis
rack_data = data[['Date', 'Sales', 'Racks']].copy()

# Convert Date to datetime format
rack_data['Date'] = pd.to_datetime(rack_data['Date'])

# Group by month and rack status to aggregate sales
rack_data['Month'] = rack_data['Date'].dt.to_period('M')
monthly_rack_data = rack_data.groupby(['Month', 'Racks']).agg({'Sales': 'sum'}).reset_index()

# Plot the impact of visually appealing racks on sales
plt.figure(figsize=(12, 6))
for rack_status in monthly_rack_data['Racks'].unique():
    rack_sales = monthly_rack_data[monthly_rack_data['Racks'] == rack_status]
    plt.plot(rack_sales['Month'].astype(str), rack_sales['Sales'], label=f'Rack Status {rack_status}')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Impact of Visually Appealing Racks on Sales')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Predict future sales using linear regression for each rack status
future_sales = []
future_dates = pd.date_range(start=rack_data['Date'].max(), periods=6, freq='MS')[1:]

for rack_status in monthly_rack_data['Racks'].unique():
    rack_sales = monthly_rack_data[monthly_rack_data['Racks'] == rack_status]
    X = np.arange(len(rack_sales)).reshape(-1, 1)
    y = rack_sales['Sales']

    model = LinearRegression()
    model.fit(X, y)

    future_X = np.arange(len(rack_sales), len(rack_sales) + 5).reshape(-1, 1)
    rack_future_sales = model.predict(future_X)

    future_sales.append((rack_status, rack_future_sales))

# Plot future forecast for each rack status
plt.figure(figsize=(12, 6))
for rack_status, sales in future_sales:
    rack_sales = monthly_rack_data[monthly_rack_data['Racks'] == rack_status]
    plt.plot(rack_sales['Month'].astype(str), rack_sales['Sales'], label=f'Past Sales - Rack Status {rack_status}')
    plt.plot(future_dates.astype(str), sales, label=f'Forecasted Sales - Rack Status {rack_status}', linestyle='--')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Sales Forecast for Next 5 Months by Rack Status')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
